```python
#Chargement des fichiers
train_df=pd.read_csv('C:/Users/Utilisateur/Documents/Projets Kaggle/Projet titanic/train.csv')
test_df=pd.read_csv('C:/Users/Utilisateur/Documents/Projets Kaggle/Projet titanic/test.csv')
```

```python
target=train_df['Survived']
inputs=train_df.drop(['Survived'],axis='columns')
inputs
```

```python
le_name=LabelEncoder()
le_sex=LabelEncoder()
l_age=LabelEncoder()
le_SibSp=LabelEncoder()
le_ticket=LabelEncoder()
le_fare=LabelEncoder()
le_cabin=LabelEncoder()
l_embarquement=LabelEncoder()
inputs['sex_n']=le_sex.fit_transform(inputs['Sex'])
inputs['Age_n']=l_age.fit_transform(inputs['Age'])
inputs['SibSp_n']=le_SibSp.fit_transform(inputs['SibSp'])
inputs['Ticket_n']=le_ticket.fit_transform(inputs['Ticket'])
inputs['Fare_n']=le_fare.fit_transform(inputs['Fare'])
inputs['Embarked_n']=l_embarquement.fit_transform(inputs['Embarked'])
inputs=inputs.drop([['Sex'],[]'Age','SibSp','Ticket','Fare'],axis='columns')


inputs
```

```python
inputs["cabin_n"]=le_cabin.fit_transform(inputs['Cabin'])

inputs['name_n']=le_name.fit_transform(inputs['Name'])
inputs
```
